```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
length = 548252
y_voice, sr = librosa.load("46_speaker_original\Test\Speaker46\Speaker46_part5.wav")
y_leaf, sr = librosa.load('Noise\leaves.wav')
y_dish, sr = librosa.load('Noise\dishwasher.wav')
y_rain, sr = librosa.load('Noise\Rain.wav')
y_room, sr = librosa.load('Noise\Room.wav')
y_theatre, sr = librosa.load('Noise\Theatre_hall.wav')
```

```python
librosa.display.waveshow(y_theatre[0:length])
```

```python
fig, ax = plt.subplots()
y = y_voice[0:length]+y_theatre[0:length]
librosa.display.waveshow(y_voice[0:length], color='r', sr=sr, ax=ax, alpha = 1, label='original audio')
librosa.display.waveshow(y[0:length], sr=sr, color='b', ax=ax, alpha = 0.7, label='noisy audio')
ax.legend()
```
